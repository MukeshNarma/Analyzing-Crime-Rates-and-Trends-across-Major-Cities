```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = {
    'Exercise Hours': [1, 2, 3, 4, 5, 6],
    'Weight Loss': [1, 2, 2.5, 3, 4, 5],
    'Diet Quality': ['Poor', 'Average', 'Good', 'Good', 'Excellent', 'Excellent']
}
df = pd.DataFrame(data)
sns.scatterplot(x='Exercise Hours', y='Weight Loss', hue='Diet Quality', data=df)
plt.title('Exercise Hours vs Weight Loss with Diet Quality')
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm
data = {
    'Teaching Method': [1, 1, 1, 0, 0, 0],
    'Test Score': [90, 85, 88, 78, 80, 74]
}
df = pd.DataFrame(data)
X = df[['Teaching Method']]
y = df['Test Score']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
import pandas as pd
import statsmodels.api as sm

# Simulated data
data = {
    'Exercise Hours': [1, 2, 3, 4, 5],
    'Weight Loss': [1, 2, 2.5, 3, 4],
    'Diet Quality': [2, 2, 1, 1, 3]
}

# Create DataFrame
df = pd.DataFrame(data)

# Model considering assumptions for causality
X = df[['Exercise Hours', 'Diet Quality']]
y = df['Weight Loss']

# Add constant term for intercept
X = sm.add_constant(X)

# Fit model
model = sm.OLS(y, X).fit()

# Print summary
print(model.summary())
```
