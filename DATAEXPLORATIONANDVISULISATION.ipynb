```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Simulated data
data = {
    'Poll Results': [45, 50, 55, 60, 65],
    'Candidate Support': [48, 53, 58, 63, 68]
}

df = pd.DataFrame(data)

# Plot the influence of Poll Results on Candidate Support
sns.scatterplot(x='Poll Results', y='Candidate Support', data=df)
plt.title('Influence of Poll Results on Candidate Support')
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm

# Sample dataset
data = {
    'Education Level': [1, 2, 3, 1, 2, 3, 1, 2, 3],
    'Years of Experience': [5, 10, 15, 4, 8, 14, 6, 12, 16],
    'Income': [50000, 60000, 70000, 55000, 65000, 75000, 52000, 63000, 72000]
}

df = pd.DataFrame(data)

# Independent variable: Education Level
# Prior variable: Years of Experience
# Dependent variable: Income

# Adding constant term for the intercept
X = df[['Education Level', 'Years of Experience']]
X = sm.add_constant(X)
y = df['Income']

# Fit the model
model = sm.OLS(y, X).fit()

# Print summary
print(model.summary())
```

```python
import pandas as pd
import statsmodels.api as sm

# Sample dataset
data = {
    'Exercise Hours': [2, 3, 1, 4, 3, 5, 2, 4, 3],
    'Diet Quality': [4, 5, 3, 5, 4, 6, 4, 5, 4],
    'Weight Loss': [2, 3, 1, 4, 3, 5, 2, 4, 3]
}

df = pd.DataFrame(data)

# Independent variable: Exercise Hours
# Intervening variable: Diet Quality
# Dependent variable: Weight Loss

# Adding constant term for the intercept
X = df[['Exercise Hours', 'Diet Quality']]
X = sm.add_constant(X)
y = df['Weight Loss']
# Fit the model
model = sm.OLS(y, X).fit()

# Print summary
print(model.summary())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Study Hours': [2, 3, 5, 7, 6, 8, 4, 5, 6],
    'Exam Score': [55, 60, 70, 80, 75, 85, 65, 70, 72]
}

df = pd.DataFrame(data)

# Scatter plot to visualize relationship
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Study Hours', y='Exam Score', data=df)
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.show()
```
